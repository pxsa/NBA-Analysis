```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import re
```

# NBA

```python
url = "https://www.basketball-reference.com/players/a/allenra02.html"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.basketball-reference.com/",
    "Connection": "keep-alive",
}

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
players_data = []

meta = soup.find('div', id='meta')
full_name = meta.find('h1').text.strip()
nationality = meta.find('span', class_='f-i').get_text(strip=True)

paragraphs = meta.find_all('p')
for p in paragraphs:
    text = p.get_text(strip=True)
    print(text)
    print('-'*50)

    if 'Career Length' in text:
        career_length = text.replace('Career Length:', "").strip()

    elif 'Position' in text:
        position = re.search(r'Position:\s*(.+)', text).group(1).strip()
        shoots = re.search(r'Shoots:\s*(\w+)', text).group(1).strip()

    elif 'College' in text:
        college = text.replace('College:', '').strip()

    elif 'High School' in text:
        high_school = text.replace('High School:', '').strip()

    elif 'Born' in text:
        birth_date = re.search(r'Born:\s*(.*?)\s*in\b', text).group(1).strip()
        born = re.search(r'in(.*)', text).group(1).strip()

    # 6-5,205lb(196cm, 92kg)
    elif 'lb' in text or 'kg' in text or 'cm' in text:
        height = re.search(r'.*?\((\d{3})cm', text).group(1).strip()
        weight = re.search(r'(\d{2,3})kg', text).group(1).strip()

player = {
    'full_name': full_name,
    'birth_date': birth_date,
    'born': born,
    'height': height,
    'weight': weight,
    'nationality': nationality,
    'position': position,
    'shoots': shoots,
    'current_team': None,
    'college': college,
    'high_school': high_school,
    'career_length': career_length,
    'games': None,
    'points': None,
    'total_rebound': None,
    'assists': None,
    'FG%': None,
    'FG3%': None,
    'FT%': None,
    'eFG%': None,
    'link': url
}

players_data.append(player)
```

```python
df = pd.DataFrame(players_data)
df.head()
```
