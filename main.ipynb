```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
headers = {"User-Agent": "Mozilla/5.0 (educational scraping practice)"}

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
# soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table", class_="table")
rows = table.find_all("tr", class_="team")
print(len(rows), "تیم پیدا شد")
```

```python
data = []
for row in rows:
    team_name = row.find("td", class_="name").text.strip()
    year = row.find("td", class_="year").text.strip()
    wins = row.find("td", class_="wins").text.strip()
    losses = row.find("td", class_="losses").text.strip()
    win_pct = row.find("td", class_="pct").text.strip()
    
    data.append({
        "team": team_name,
        "year": year,
        "wins": wins,
        "losses": losses,
        "win_pct": win_pct
    })

for d in data[:5]:
    print(d)
```

```python
all_data = []
for page in range(1, 25):  # این سایت 24 صفحه داره
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.text, "lxml")
    rows = soup.find_all("tr", class_="team")
    
    for row in rows:
        all_data.append({
            "team": row.find("td", class_="name").text.strip(),
            "year": row.find("td", class_="year").text.strip(),
            "wins": row.find("td", class_="wins").text.strip(),
            "losses": row.find("td", class_="losses").text.strip(),
        })
    
    # time.sleep(2)

print(f"مجموع رکوردها: {len(all_data)}")
```
