### cleaning and standardization Players raw data

```python
players = players_raw.drop(columns=["Unnamed: 0"])
```

```python
players["birth_date"].head(20).apply(repr)
```

```python
## converting birthdate format to datetime 
```

```python
players["birth_date"] = (
    players["birth_date"]
    .str.replace(r"\s*,\s*", ", ", regex=True)
)

players["birth_date"] = pd.to_datetime(
    players["birth_date"],
    format="%B %d, %Y",
    errors="coerce"
)
```

```python
players.head(10)
```

```python
## position column standardization
```

```python
position_map = {
    "Point Guard": "PG",
    "Shooting Guard": "SG",
    "Small Forward": "SF",
    "Power Forward": "PF",
    "Center": "C",
    "Guard": "G",
    "Forward": "F"
}
```

```python
def normalize_positions(pos):

    if pd.isna(pos):
        return None

    parts = re.split(r",| and |/", pos)

    positions = []

    for part in parts:
        part = part.strip()

        if part in position_map:
            positions.append(position_map[part])

    return sorted(set(positions))
```

```python
players["position"] = players["position"].apply(normalize_positions)
```

```python
## shoots column standardization
```

```python
shoot_map = {"Right": "Right",
             "Left": "Left",
             "LeftRight": "Both",
             "RightLeft": "Both"}

players["shoots"] = players["shoots"].replace(shoot_map)
```

### college DataFrame

```python
def parse_colleges(college):
    if pd.isna(college):
        return []

    # rmeving college 
    college = college.replace("Colleges:", "")

    
    colleges = college.split(",")

    # removing extra soaces
    colleges = [c.strip() for c in colleges]

    # removing empty strings
    colleges = [c for c in colleges if c]

    return colleges
```

```python
all_colleges = []

for colleges in players["college"]:
    all_colleges.extend(parse_colleges(colleges))
```

```python
college_df = pd.DataFrame({
    "college_name": sorted(set(all_colleges))})
```

```python
college_df.insert(
    0,
    "college_id",
    range(1, len(college_df)+1))
```

```python
college_df.head()
```

```python
college_df.shape
```

### player_college DataFrame

```python
# mapping college to each player
```

```python
college_map = dict(
    zip(
        college_df["college_name"],
        college_df["college_id"]))
```

```python
player_college_rows = []

for _, row in players.iterrows():

    colleges = parse_colleges(row["college"])

    for college in colleges:

        player_college_rows.append({
            "player_id": row["player_id"],
            "college_id": college_map[college]})
```
